```python
RECORDS = (0, 2, 4, 5)
COURSES = ("LC Men", "LC Women", "SC Men", "SC Women")

records = {}
for table, course in zip(RECORDS, COURSES):
   #Recreate my table
   df = tables[table][["Event", "Time"]]
   #Get rid of relay
   df = df[~df["Event"].str.contains("relay")]
   # As df returns a table with index, and columns
   # We need to reshape the dataframe to use "Event"
   # as the index value, so it becomes without the index
   df = df.set_index("Event")
   records[course] = df.to_dict()["Time"]
  
records
```
